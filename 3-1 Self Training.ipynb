```python
pairs_n = pairs[pairs["player1"].isin(["Carlos Alcaraz", "Daniil Medvedev", "Laslo Djere"]) & pairs["player2"].isin(["Nicolas Jarry", "Marton Fucsovics", "Stefanos Tsitsipas"])]
pairs_n
```

```python
for index, row in pairs_n.iterrows():
    
    p1 = row["player1"]
    p2 = row["player2"]

    u = tr.ModelUnit(p1, p2, data)

    u.preprocess()
    u.prefilter()

    u.calc_torque(players["Inertia"])
    u.train(inputs)
    #u.show_params()
    u.construct_momentum()
    #u.compare_constructed()
    #vs.save(f"flow/4-2-Flow-{p1[0]}")
    u.construct_energy()
    #u.compare_performance()
    #vs.save(f"flow/4-2-Perform-{p1[0]}")
    u.summary()
    plt.tight_layout()
    vs.save(f"flow/4-2-Combine-{p1[0]}")
```

```python

import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import numpy as np

# Generate sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a plot
plt.plot(x, y1, label='Sin(x)', color='blue')
plt.plot(x, y2, label='Cos(x)', color='red')

# Create a yellow rectangle for the first legend
yellow_rectangle = Rectangle((0, 0), 1, 1, fc='yellow', edgecolor='none')

# First legend with the yellow rectangle
legend1 = plt.legend([yellow_rectangle], ['Custom Legend'], loc='upper right')

# Second legend for the lines
legend2 = plt.legend(loc='upper left')

# Add both legends to the plot
plt.gca().add_artist(legend1)

# Show the plot
plt.show()
```
